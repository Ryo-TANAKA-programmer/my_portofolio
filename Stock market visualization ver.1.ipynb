```python
# Import necessary libraries
import yfinance as yf #yahoo finance library
import matplotlib.pyplot as plt
%matplotlib inline

# Obtain the historical price data from DOW
ticker = ['TSLA','LCID','TM']
data = yf.download(ticker, start = '2022-01-01', end = '2023-01-01')
data['Adj Close'].plot()
# adjusted closed price of stock
plt.title('Stock Price of EV Manufacturers')
plt.savefig('stock.png')
```

```python
import yfinance as yf
import streamlit as st
import pandas as pd

st.write("""
# Simple Stock Price App
shown below are the stock closing price and volume of Google
""")

# https://towardsdatascience.com/how-to-get-stock-data-using-python-c0de1df17e75

#define the ticker symbol
tickerSymbol = 'GOOGL'

#get data of the ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices of the year

tickerDf = tickerData.history(period='1d', start='2015-5-31',end='2020-5-31')

#Open  High  Low Close  Volume  Dividends  Stock splits

st.line_chart(tickerDf.Close)
st.line_chart(tickerDf.Volume)
```
